Import library

```python
#Import Library

import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
#Membaca dan Melihat Dataset

data = pd.read_csv('Attrition.csv')
data
```

```python
#Melihat Informasi dari dataset

data.info()
```

```python
#Mengambil atribut dari dataset yang akan digunakan

data = data[['Age', 'Gender', 'BusinessTravel', 'Education', 'Department', 'DailyRate','DistanceFromHome','JobInvolvement','TotalWorkingYears','Attrition']]
data.head()
```

```python
#Mengganti Nama Atribut yang akan digunakan

data = data.rename(columns={'Age':'Umur', 'Gender':'Kelamin', 'BusinessTravel':'BisnisPerjalanan', 'Education':'Pendidikan', 'Department':'Departemen', 'DailyRate':'RateHarian','DistanceFromHome':'JarakDariRumah','JobInvolvement':'KepuasanKerja','TotalWorkingYears':'TotalBekerjaSetahun'})
data.head()
```

# DATA CLEANING

## **1. Nilai Null**

Mengganti nilai null dengan mean

```python
#Mencari Nilai Null

data.isna().sum()
```

```python
#Melihat nilai mean dari dataset

data.mean()
```

```python
#Mengganti value null menjadi nilai mean

data = data.fillna(data.mean())
```

```python
data.isna().sum()
```

## **2. Nilai Duplikat**

Menghapus Nilai Duplikat

```python
#Melihat Jumlah nilai duplikat

data.duplicated().sum()
```

```python
#Menampilkan Nilai Duplikat

data[data.duplicated()]
```

```python
#Menghapus Nilai Duplikat

data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

## **3. Mengganti Tipe data salah satu attribute angka**

Mengganti Atribut float menjadi int

```python
data.info()
```

```python
#Mengganti tipe data atribut umur dari float menjadi int

data['Umur'] =  data['Umur'].astype(int)
```

```python
data.info()
```

```python
data.head()
```

# DATA TRANSFORMING

## **1. Normalisasi**

```python
#Menormalisasikan data menggunakan min max Scaller

minmax = data.copy()
mms = MinMaxScaler()
minmax_scaled = mms.fit_transform(minmax[['JarakDariRumah','KepuasanKerja','TotalBekerjaSetahun']])
data_normalization = pd.DataFrame(minmax_scaled)
```

```python
data_normalization.head()
```

## **Standarisasi**

```python
#Melakukan standarisasi

standar = data.copy()
ss = StandardScaler()
Scaled_data = ss.fit_transform(standar[['Umur','Pendidikan']])
```

```python
print("Nilai Data Sebelum scaling : ")
print(data[['Umur','Pendidikan']])
print("Nilai standar deviasi: ", np.std(data))

print("\n Nilai Data setelah scaling : ")
print(Scaled_data)
print("Nilai standar deviasi: ", np.std(Scaled_data))
```

# ENCODING

# DATA SPLIT

## **Membagi Dataset Menjadi Training set dan Testing set**

```python
#Mengambil nilai x dan y, dan menjadikan atribut Attrition menjadi Label

x = data.drop('Attrition', axis=1)
y = data['Attrition']
```

```python
#Membagi dataset manjadi train set dan test set

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42) 
```
